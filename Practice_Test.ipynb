Write a Python program that:

Takes a sentence as input from the user.

Reverses only the words in the sentence

that have an odd number of characters.

Keeps the order of the words the same.

Prints the updated sentence.
Example:



Input: Python is amazing and fun

Output: nohtyP is gnizama and fun



```python
#input the sentence
input_sentence = input("Enter a sentence: ")
```

```python
def reverse_odd_length_words(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Reverse words with odd number of characters
    updated_words = []
    for word in words:
        if len(word) % 2 != 0:
            updated_words.append(word[::-1])
        else:
            updated_words.append(word)

    # Join the words back into a sentence
    return ' '.join(updated_words)

# Take input from the user
user_input = input("Enter a sentence: ")

# Process and print the updated sentence
result = reverse_odd_length_words(user_input)
print("Updated sentence:", result)
```
